Helper Predict Function -  function that would get a single data point value. Where most of the computation is at.

```python
import numpy as np
from collections import Counter

def euclidean_distance(x1, x2): #Given two arrays, calculate the distance
    distance = np.sqrt(np.sum( (x1-x2)**2 )) #Apparently we only need to do the x portion
    return distance

class KNN:
    def __init__(self, k=3): #k=3 is needed for [: self.k]
        self.k = k
        
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        
    def predict(self, X): #Main Predict Function
        predictions = [ self._predict(x) for x in X] #return helper function, for each examples in the dataset.
        return predictions
        
  
    def _predict(self, x): #Helper Function
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train] 
        #x = x that is passed to the function, x_train = each value in x_train (which is self.X_train)
        
        #If we pass X_test here, we get the distance of X_test and X_Train
        
        #get the closest k
        k_indices = np.argsort(distances)[:self.k] #Use np.argsort and then get the first k -- the closest neighbors
        k_nearest_labels = [self.y_train[i] for i in k_indices] #for index in k_indices, plot it against y_train, to find the nearest label.
        
        #Determine the label with majority vote
        most_common = Counter(k_nearest_labels).most_common() # This function returns the most common label
        return most_common  
```

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
#from KNN import KNN #Usually we add this if it's a separate module, but since it's in jupyter no need
#from the KNN.py file, import the class KNN


cmap = ListedColormap( ['#FF0000', '#00FF00', '#0000FF'] )

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =1234)


plt.figure()
plt.scatter(X[: ,2] , X[: ,3] , c=y, cmap=cmap, edgecolor = 'k', s= 20)

#Add Labels
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Petal Length vs Petal Width')

#Add Legend - FROM CHAT GPT
classes = iris.target_names
class_colours = ['#FF0000', '#00FF00', '#0000FF']
recs = []
for i in range(len(class_colours)):
    recs.append(mpatches.Rectangle((0,0),1,1,fc=class_colours[i]))
plt.legend(recs, classes, loc='upper left')
#Enable Grid
plt.grid(True)

#Call Command
plt.show()

#Call all the functions, EXCEPT the helper function
clf = KNN(k=5)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

acc = np.sum(predictions == y_test) / len(y_test)
```
