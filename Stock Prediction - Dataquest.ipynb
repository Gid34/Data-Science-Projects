```python
## Backtesting
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])  #same preds as next line
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#We have previously defined model, predictors as the following

# model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state=1)

# predictors = ["Close", "Volume", "Open", "High", "Low"]



def backtest(data, model, predictors, start = 2500, step = 250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy() 
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model) #Current Predictions of that state
        all_predictions.append(predictions)
        
    return pd.concat(all_predictions)
    
    
```

```python
backtest_result = backtest(sp500, model, predictors)

#even if data was not defined, the arg sp500 was passed for the code to be executed
```

```python
horizons = [2,5,60,250,1000]

#calculate mean close price in the last 2 days, 5 days ,60 days, 1 year , last 4 yrs

new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"] 
    #Today's Close / (Day 1 + Day2) -> Today's Close / (Day 1 + Day2 + Day3) 
    
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"] 
    #Shift forward and find the rolling sum of the target

    new_predictors += [ratio_column, trend_column] 
```

# Improving the Model

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state=1)
#We've only changed the n_estimators and min_samples_split in this run, WE KEPT SAME random state
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    
        #New revision of code below
    preds = model.predict_proba(test[predictors]) [:, 1] 
    preds[preds >= 0.6] = 1 #custom threshold: if probability is 60 or above, return 1 or stock price up
    preds[preds < 0.6] = 0 #for probability of less than 60, return 0 or stock price down
        #End of revision
    
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```
