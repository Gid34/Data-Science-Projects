```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

points = {
    "blue" : [ [2,4], [1,3], [2,3], [3,2], [2,1] ],
    "red" : [ [5,6], [4,5], [4,6], [6,6], [5,4] ]
}

new_point = [3,3]

def euclidean_distance(p, q):
    return np.sqrt(np.sum ( (np.array(p) - np.array(q)) **2) )
    
    
class KNN:
    def __init__(self, k=3):
        self.k = k
        self.point = None
        
    def fit(self, points):
        self.points = points
    
    def predict(self, new_point):
        distances = []
        
        for category in self.points: #for blue or red in self.points
            for point in self.points[category]:
                distance = euclidean_distance(point, new_point) #Distance from current point and new point
                distances.append([distance, category]) #Append to the list two values: the distance and the category
        
        categories = [category[1] for category in sorted(distances)[:self.k]] #Return the category, AFTER the value is sorted up to the 3 nearest neighbors
        #Ex. [blue, blue, red]
        
        result = Counter(categories).most_common(1)[0][0] #Return the most common from the list
        return result
    
    
clf = KNN()
clf.fit(points)
print(clf.predict(new_point))
```

```python
#Visualize

ax = plt.subplot()
ax.grid(True, color = "#323232")

#ax.set_facecolor("black")
# ax.figure.set_facecolor("#121212")

ax.tick_params(axis = "x", color = "white")
ax.tick_params(axis = "y", color = "white")

for point in points['blue']:
    ax.scatter(point[0], point[1], color = "#104DCA", s=60)

for point in points['red']:
    ax.scatter(point[0], point[1], color = "#FF0000", s=60)

#Code to return the color
new_class = clf.predict(new_point)
color = "#FF0000" if new_class == "red" else "#104DCA"
ax.scatter(new_point[0], new_point[1], color = color, marker = "*", s=200, zorder=100)


#Draw a dashed line for each point
for point in points['blue']:
    ax.plot([new_point[0], point[0]],[new_point[1], point[1]], color = "#104DCA", linestyle = "--", linewidth = 1 )
    

for point in points['red']:
    ax.plot([new_point[0], point[0]],[new_point[1], point[1]], color = "#FF0000", linestyle = "--", linewidth = 1 )
    
plt.show()
```

```python
points = {
    "blue" : [ [2,4,3], [1,3,5], [2,3,1], [3,2,3], [2,1,6] ],
    "red" : [ [5,6,5], [4,5,2], [4,6,1], [6,6,1], [5,4,6], [10,10,4] ]
}

# new_point = [3,3,100] #Result blue
new_point = [3,100,100] #Result red

def euclidean_distance(p, q):
    return np.sqrt(np.sum ( (np.array(p) - np.array(q)) **2) )
    
    
class KNN:
    def __init__(self, k=3):
        self.k = k
        self.point = None
        
    def fit(self, points):
        self.points = points
    
    def predict(self, new_point):
        distances = []
        
        for category in self.points: #for blue or red in self.points
            for point in self.points[category]:
                distance = euclidean_distance(point, new_point) #Distance from current point and new point
                distances.append([distance, category]) #Append to the list two values: the distance and the category
        
        categories = [category[1] for category in sorted(distances)[:self.k]] #Return the category, AFTER the value is sorted up to the 3 nearest neighbors
        #Ex. [blue, blue, red]
        
        result = Counter(categories).most_common(1)[0][0] #Return the most common from the list
        return result
    
    
clf = KNN()
clf.fit(points)
print(clf.predict(new_point))

#Data Visualization
fig = plt.figure(figsize = (15, 12))
ax = fig.add_subplot(projection = "3d") #To enable 3D
ax.grid(True, color = "#323232")

#ax.set_facecolor("black")
# ax.figure.set_facecolor("#121212")

ax.tick_params(axis = "x", color = "white")
ax.tick_params(axis = "y", color = "white")

for point in points['blue']:
    ax.scatter(point[0], point[1], point[2], color = "#104DCA", s=60)

for point in points['red']:
    ax.scatter(point[0], point[1], point[2], color = "#FF0000", s=60)

#Code to return the color
new_class = clf.predict(new_point)
color = "#FF0000" if new_class == "red" else "#104DCA"
ax.scatter(new_point[0], new_point[1], new_point[2], color = color, marker = "*", s=200, zorder=100)


#Draw a dashed line for each point
for point in points['blue']:
    ax.plot([new_point[0], point[0]], [new_point[1], point[1]], [new_point[2], point[2]], color = "#104DCA", linestyle = "--", linewidth = 1 )
    

for point in points['red']:
    ax.plot([new_point[0], point[0]], [new_point[1], point[1]], [new_point[2], point[2]], color = "#FF0000", linestyle = "--", linewidth = 1 )
    
plt.show()
```
