## First Run of GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

#Specify a parameter grid with the following hyperparameters

forest = RandomForestRegressor()

X_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

param_grid = {
    "n_estimators": [3,10,30],
    "max_features": [2,4,6,8],
    
}

#How the cross validation works -> n_estimator x max_features
#Ex. (3x2), (3x4), ... exhaust list before (10x2), (10x4)...until finish

#GridSearchCV(estimator, param_grid, *, scoring=None, n_jobs=None, refit=True, cv=None, verbose=0, pre_dispatch='2*n_jobs', error_score=nan, return_train_score=False)

#estimator - object to be passed on
#param_grid - dictionary or list of dictionaries


grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = "neg_mean_squared_error",
                           return_train_score = True)

grid_search.fit(X_train_s, y_train)
```

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

```python
best_forest = grid_search.best_estimator_
```

```python
#GridSearchCV Score
best_forest.score(X_test_s, y_test)
```

Apparently once you fit the train data, the test data is also fitted?

## Adjust the data for 2nd GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [100,200,300],
    "max_features": [2,4,6,8],
    
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = "neg_mean_squared_error",
                           return_train_score = True)

grid_search.fit(X_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
#GridSearchCV Score2
best_forest.score(X_test_s, y_test)
```

```python
#Finally worked!
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [100,200,300],
    "min_samples_split": [2,4],
    "max_depth": [None, 4, 8]
    
}



grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = "neg_mean_squared_error",
                           return_train_score = True)

grid_search.fit(X_train_s, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test_s, y_test)
```
